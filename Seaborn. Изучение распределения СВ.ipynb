```python
import requests
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%%time
L = []
url = 'https://math-praktikum-stepik.herokuapp.com/api/assignment/a368500_10/'
for _ in range(1000): 
    r = requests.get(url).text
    L.append(float(r))
```

```python
L = np.array(L)
m = L.mean()
s = L.std(ddof=1)
print(f'Выборочное среднее: {m:.2f}')
print(f'Выборочное стандартное отклонение: {s:.2f}')
```

```python
plt.figure(figsize=(12, 7))
sns.set_context('notebook', font_scale=1.5)
ax = sns.histplot(data=L, bins=25, kde=True, fill=False, color='gray')
x, y = ax.lines[0].get_data()
plt.vlines(m, color='black', ls='--', ymax=40, ymin=0, label='$\\bar x$')
ax.fill_between(x, y, where=((x >= m - s) & (x <= m + s)), label='$\\bar x \\pm \\sigma$', color='yellow')
ax.fill_between(x, y, where=(((x >= m - 2 * s) & (x <= m + 2 * s))), alpha=0.3, color='blue',
               label='$\\bar x \\pm 2 \\sigma$')
plt.legend()
plt.title('Гистограмма распределения случайной величины')
plt.xlabel('$x$')
```

По рисунку можно заметить, что распределение является бимодальным, т. е. имеет несколько пиков. <br> Так, первый из пиков приходится на интервал (0.43, 1]. Второй -- на интервал (5.60, 6.18].
