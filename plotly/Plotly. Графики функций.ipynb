Постройте графики 2 функций:
- f(x) = sin(x)
- g(x) = 1 - cos(x)

Для x на отрезке [-2$\pi$; 2$\pi$]

Отметьте отдельными маркерами точки пересечения графиков. Оба графика должны проходить через точку пересечения при любом приближении (т.е. найдите точки пересечения с абсолютной точностью и добавьте как значения аргумента, так и значение функции в список по которому строится кривая).

*Примечание1. Для поиска точек пересечения имеет смысл аналитически решить тригонометрическое уравнение*

    sin(x) = 1 - cos(x)
    
Примечание2. Чтобы решить это уравнение вам могут потребоваться формулы:
$$sin^{2}\alpha =\frac{1-cos2\alpha }{2}$$

$$sin 2\alpha  = 2 \cdot  sin \alpha \cdot  cos \alpha $$

* Подпишите оси и выведите заголовок

* Выведите осевые линии

```python
import numpy as np
import plotly.graph_objs as go
```

```python
def f(x): 
    return np.sin(x)

def g(x): 
    return 1 - np.cos(x)

x = np.arange(-2 * np.pi, 2 * np.pi + 1, np.pi/16)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=f(x), name='f(x) = sin(x)'))
fig.add_trace(go.Scatter(x=x, y=g(x), name='g(x) = 1 - cos(x)'))
fig.update_layout(xaxis=dict(tickmode='array', tickvals=np.arange(-2 * np.pi, 2 * np.pi + np.pi/2, np.pi/2), 
                            ticktext=['$-2\pi$', '$-3\pi / 2$', '$-\pi$', '$-\pi / 2$', '0', 
                                     '$\pi / 2$', '$\pi$', '$3\pi / 2$', '$2\pi$']))

intersect_x = [-2 * np.pi, -3 * np.pi / 2, 0, np.pi / 2, 2 * np.pi / 1]
fig.add_trace(go.Scatter(x=intersect_x, y=f(intersect_x), mode='markers', marker=dict(size=10), 
                        name='Точка пересечения'))
fig.update_layout(xaxis_title='$x$', yaxis_title='$f(x), g(x)$', 
                  title=dict(text='Графики функций f(x), g(x)', xanchor='center', x=0.5, y=0.9))
fig.update_xaxes(zeroline=True, zerolinecolor='lightgray')
fig.update_yaxes(zeroline=True, zerolinecolor='lightgray')
fig.show()
```
