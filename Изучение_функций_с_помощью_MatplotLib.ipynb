## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
def f(x, alpha, beta): 
    return (x**beta + alpha**beta) / x**beta
```

```python
old_settings = np.seterr(all='warn')
np.seterr(all='ignore')
```

```python
x = np.linspace(-6, 6, 111)

plt.figure(figsize=(16, 9))
plt.plot(x, f(x, alpha=1, beta=1), label='$\\alpha=1, \\beta=1$', color='seagreen')
plt.plot(x, f(x, alpha=2, beta=1), label='$\\alpha=2, \\beta=1$', color='orangered')
plt.plot(x, f(x, alpha=1, beta=2), label='$\\alpha=1, \\beta=2$', color='darkslateblue')

plt.ylabel('$f(x)$', fontsize=16)
plt.xlabel('$x$', fontsize=16)
plt.title('График функции $f(x) = \\frac{x^\\beta + \\alpha^\\beta}{x^\\beta}$ при различных $\\alpha$ и $\\beta$', 
         fontsize=18)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend(fontsize='xx-large', title='Параметры функции', title_fontsize=16)
plt.grid()
plt.savefig('task1.svg')
```

По графику видно, что исследуемая функция имеет точку разрыва при $x = 0$ (т.к. $x^\beta$ является знаменателем).<br>

Таким образом, функция имеет вертикальную асимптоту $x = 0$.<br>

При четных $\beta$ $f(x) > 0$ и симметрична относительно оси ординат.<br>

Функция пересекает ось ординат при $x = -\alpha$ при нечетных $\beta$.<br>

### Часть 2

Построить в общих осях графики для $x>0$.

На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:

- для малых x

- для больших x

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
x = np.linspace(0.05, 10, 500)

fig = plt.figure(figsize=(16, 9))

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, f(x, alpha=1, beta=1), label='$\\alpha=1, \\beta=1$', color='seagreen')
axes1.plot(x, f(x, alpha=2, beta=1), label='$\\alpha=2, \\beta=1$', color='orangered')
axes1.plot(x, f(x, alpha=1, beta=2), label='$\\alpha=1, \\beta=2$', color='darkslateblue')

axes1.set_title(
    '''График функции $f(x) = \\frac{x^\\beta + \\alpha^\\beta}{x^\\beta}$ при различных $\\alpha$ и $\\beta$, $x > 0$''',
    fontsize=20)
axes1.legend(fontsize=16, title='Параметры функции', title_fontsize=14)
axes1.set_xlabel('$x$', fontsize=16)
axes1.set_ylabel('$f(x)$', fontsize=16)
axes1.tick_params(labelsize=14, axis='both')


#врезка графика фукнции при x->0
x_zero = np.linspace(0.25, 3, 1000)
axes2 = fig.add_axes([0.1, 0.3, 0.4, 0.4])
axes2.plot(x_zero, f(x_zero, alpha=1, beta=1), color='seagreen')
axes2.plot(x_zero, f(x_zero, alpha=2, beta=1), color='orangered')
axes2.plot(x_zero, f(x_zero, alpha=1, beta=2), color='darkslateblue')
axes2.set_title('$f(x)$ при малых $x$', fontsize=16)
axes2.set_xlabel('$x$', fontsize=14)
axes2.set_ylabel('$f(x)$', fontsize=14)
axes2.tick_params(labelsize=11, axis='both')
axes2.grid()

#врезка графика функции при x->+inf
x_inf = np.linspace(1000, 5000, 4000)
axes3 = fig.add_axes([0.55, 0.3, 0.4, 0.4])
axes3.plot(x_inf, f(x_inf, alpha=1, beta=1), color='seagreen')
axes3.plot(x_inf, f(x_inf, alpha=2, beta=1), color='orangered')
axes3.plot(x_inf, f(x_inf, alpha=1, beta=2), color='darkslateblue')
axes3.set_title('$f(x)$ при больших $x$', fontsize=16)
axes3.set_xlabel('$x$', fontsize=14)
axes3.tick_params(labelsize=11, axis='both')

plt.savefig('task2.svg')
```

Рассмотрим случай когда $x, \alpha, \beta > 0$. 

$\large f(x) = \frac{x^\beta + \alpha^\beta}{x^\beta} = 1 + \frac{\alpha^\beta}{x^\beta}$ <br>

Тогда $\large \lim_{x \to +0} f(x) = +\infty$ (т. е. возрастает).<br>

$ \large \lim_{x \to +\infty} f(x) = 1$ (сверху, т. е. убывает).

При этом так как $x^2$ стремится к $\infty$ быстрее, то чем больше $\beta$ и чем меньше $\alpha$, тем быстрее $f(x)$ стремится к 1.

Также видно, что $f(x)$ при $\alpha = 1, \beta = 2$ пересекает две другие функции. 

```python
print(f(5 * 10**3, alpha=1, beta=2)) 
print(f(5 * 10**3, alpha=1, beta=1))
print(f(5 * 10**3, alpha=2, beta=1))
```

### Часть 3

Построить в общих осях графики для $x<0$.

На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к $-\infty$.

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. Так же нанесите на графики прямую f(x) = 0.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
x = np.linspace(-50, -1, 1000)

fig = plt.figure(figsize=(16, 9))

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, f(x, alpha=1, beta=1), label='$\\alpha=1, \\beta=1$', color='seagreen')
axes1.plot(x, f(x, alpha=2, beta=1), label='$\\alpha=2, \\beta=1$', color='orangered')
axes1.plot(x, f(x, alpha=1, beta=2), label='$\\alpha=1, \\beta=2$', color='darkslateblue')

axes1.set_title(
    'График функции $f(x) = \\frac{x^\\beta + \\alpha^\\beta}{x^\\beta}$ при различных $\\alpha$ и $\\beta$, $x < 0$',
    fontsize=20)
axes1.legend(fontsize=16, title='Параметры функции', title_fontsize=14)
axes1.set_xlabel('$x$', fontsize=16)
axes1.set_ylabel('$f(x)$', fontsize=16)
axes1.tick_params(labelsize=14, axis='both')
axes1.axhline(color='lightgray', linewidth=1)
axes1.axhline(y=1, color='lightgray', linewidth=1)

#врезка графика фукнции при x->0
x_zero = np.linspace(-2, -0.25, 100)
axes2 = fig.add_axes([0.15, 0.1, 0.4, 0.4])
axes2.plot(x_zero, f(x_zero, alpha=1, beta=1), color='seagreen')
axes2.plot(x_zero, f(x_zero, alpha=2, beta=1), color='orangered')
axes2.plot(x_zero, f(x_zero, alpha=1, beta=2), color='darkslateblue')
axes2.axhline(color='lightgray', linewidth=1)

axes2.set_title('$f(x)$ при малых $x$', fontsize=16)
axes2.set_xlabel('$x$', fontsize=14)
axes2.set_ylabel('$f(x)$', fontsize=14)
axes2.tick_params(labelsize=11, axis='both')



plt.savefig('task3.svg')
```

Рассмотрим случай, когда $\alpha, \beta > 0, x < 0$ <br>

Как упоминалось ранее, при $\beta = 2$ (т. е. четных) $f(x) > 0$.

Так же как и для $x > 0$, $\large \lim_{x \to -0} f(x) = +\infty$ (при четных $\beta$). (т. е. возрастает)

При нечетных $\beta$ $\large \lim_{x \to -0} f(x) = -\infty$ (т. е. убывает).

Для четных $\beta$ $\large \lim_{x \to -\infty} f(x) = 1$ (сверху, т. е. убывает, так же как и для x > 0). <br>

Для нечетных $\beta$ $\large \lim_{x \to -\infty} f(x) = 1$ (но снизу,т. е. возрастает).

Кроме того, видно, что два графика (где $\beta = 1$) пересекают ось ОХ, т. е. функция принимает значение, равное 0. <br>

Для $f(x)$, где $\alpha = 1$ функция принимает нулевое значение в точке -1. Для $f(x)$, где $\alpha = 2$ это точка равна -2.

Таким образом, для нечетных $\beta$ и x < 0 функция будет пересекать ось OX в точке x = -$\alpha$.<br>

Между собой графики не пересекаются.

## Задача №2

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0$

- $\alpha = 0, \beta = 0$

- $\alpha = 0, \beta = 1$

- $\alpha = 0, \beta = 2$

- $\alpha = 1, \beta = -1$

- $\alpha = 2, \beta = -1$

Сделайте выводы о поведении графиков

```python
plt.figure(figsize=(16, 9))

x = np.linspace(-10, 10, 101)
plt.plot(x, f(x, alpha=1, beta=0), label='$\\alpha=1, \\beta=0$', linewidth=5) #в python 0**0 = 1
plt.plot(x, f(x, alpha=0, beta=0), label='$\\alpha=0, \\beta=0$', color='yellow', linestyle='--')
plt.plot(x, f(x, alpha=0, beta=1), label='$\\alpha=0, \\beta=1$', linewidth=5, color='black')
plt.plot(x, f(x, alpha=0, beta=2), label='$\\alpha=0, \\beta=2$', color='red', linestyle='--')
plt.plot(x, f(x, alpha=1, beta=-1), label='$\\alpha=1, \\beta=-1$')
plt.plot(x, f(x, alpha=2, beta=-1), label='$\\alpha=2, \\beta=-1$')

plt.legend(fontsize=14)
plt.title('График функции $f(x) = \\frac{x^\\beta + \\alpha^\\beta}{x^\\beta}$ при различных $\\alpha$ и $\\beta$', fontsize=18)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.tick_params(labelsize=14, axis='both')
plt.grid()
```

```python
print(0**0)
print(np.power(0, 0))
```

Заметим, что если хотя бы один из аргументов $\alpha, \beta$ принимает нулевое значение, то функция принимает константное значение и параллельна оси OX (первая производная в таком случае равна 0). <br>

В частности, если $\beta = 0$, то $f(x) = 2$ (при x <> 0), т. к. число в степени 0 равно 1. <br>

Если $\alpha = 0$, то $f(x) = 1$. <br>

Если $\beta$ < 0, то функция приобретает вид степенной функции $f(x) = 1 + \alpha^{-\beta}x^\beta$. В соответствии со значением $\beta$ (по модулю) функция приобретает различный вид: 1 - прямая, 2 - парабола, 3 - кубическая функция и т. д. $\alpha$ выступает в качестве коэффициента, например, при $\beta = -1$, прямая имеет коэффициент $\frac{1}{\alpha}$
